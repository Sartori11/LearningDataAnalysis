<h1>excel com python</h1>


```python
import pandas as pd
import numpy as np
```

```python
# df = pd.read_csv()
# df = pd.read_json()
df = pd.read_excel('DadosVendaCarros.xlsx')
```

```python
df.head() # retorna as 5 primeiras linhas pode receber parametro
```

```python
df.tail() # 5 ultimas linhas pode receber parametro
```

```python
headers = ["EmissãoNotaFiscal","Marca","Local","Venda","Custo","Desconto","CustoDeEntrega","CustoDeMaoDeObra","Cliente","Model","Color","Year"]
df.columns = headers # TROCANDO NOME DAS COLUNAS
df.head(5)
```

```python
df1 = df.replace('Rolls Royce','Royce')
df1.head(5) # trocando valores  em todas as linhas passando parametros
```

```python
df1=df.replace(0.0,np.NaN) # trocar valores de 0 por Nan para que a func dropna possa remover esses valores

df=df1.dropna(subset=["Desconto"], axis=0)  #Dropar valores faltando na coluna "nome_da_coluna" as
df.head(5)
```

```python
df.to_csv("DadosVendaCarros.csv", index=False) #  salvar o arquivo como csv - index = False means the row names will not be written.

# df.to_json / df.to_sql / df.to_excel / df.to_hdf outras opções
```
